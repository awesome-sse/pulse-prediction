```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('video1.MOV')
images = []
if (cap.isOpened() == False):
    print("Error opening video stream or file")
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        cv2.imshow('Frame',frame)
        images += [frame]
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
